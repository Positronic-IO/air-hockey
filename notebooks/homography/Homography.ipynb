```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
#Read in the starting image
raw_img = cv2.imread("Center.jpg")
plt.imshow(raw_img)
```

```python
#Display the 3 corners
disp_img = raw_img.copy()
def mark_point(img, x, y, radius = 5):
    img[y-radius:y+radius, x-radius:x+radius] = (0, 255, 0)
mark_point(disp_img, 700, 210)
mark_point(disp_img, 1160, 190)
mark_point(disp_img, 1600, 870)
mark_point(disp_img, 260, 900)

plt.figure(figsize=(16,16))
plt.imshow(disp_img)
print(disp_img.shape)
```

```python
import time
start = time.time()
# x, y
pts_src = np.array(
    [
        [700, 210],
        [1160, 190],
        [1600, 870],
        [260, 900]
    ]
)
pts_dst = np.array(
    [
        [0, 699],
        [399, 699],
        [399, 0],
        [0, 0]        
    ]
)
h, status = cv2.findHomography(pts_src, pts_dst)

warped = cv2.warpPerspective(raw_img, h, (400, 700))
print(time.time() - start)
plt.imshow(warped)
```

```python
#Read in the starting image
raw_img = cv2.imread("Low.jpg")
plt.imshow(raw_img)
```

```python
#Display the 3 corners
disp_img = raw_img.copy()
def mark_point(img, x, y, radius = 5):
    img[y-radius:y+radius, x-radius:x+radius] = (0, 0, 255)
mark_point(disp_img, 670, 240)
mark_point(disp_img, 1160, 235)
mark_point(disp_img, 1890, 800)
mark_point(disp_img, 20, 820)

plt.figure(figsize=(16,16))
plt.imshow(disp_img)
print(disp_img.shape)
```

```python
import time
start = time.time()
# x, y
pts_src = np.array(
    [
        [670, 240],
        [1160, 235],
        [1890, 800],
        [20, 820]
    ]
)
pts_dst = np.array(
    [
        [0, 699],
        [399, 699],
        [399, 0],
        [0, 0]         
    ]
)
h, status = cv2.findHomography(pts_src, pts_dst)

warped = cv2.warpPerspective(raw_img, h, (400, 700))
plt.figure(figsize=(16, 16))
print(time.time() - start)
plt.imshow(warped)
```
